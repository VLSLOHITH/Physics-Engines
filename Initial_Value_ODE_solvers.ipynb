```
#heuns method for second order differential equations
from pylab import plot,show,grid,scatter
from math import cos
from copy import deepcopy
n=350 #range of x we are plotting
x=[0.0 for i in range(n)] #y cordinate
u=deepcopy(x)#position got by eulers method
t=[0.1*i for i in range(n)]#x cordinate
v=deepcopy(x)#slope of the equation
a=deepcopy(x)#acceleration
k=deepcopy(x)#velocity we got by eulars method
l=deepcopy(x)# slope at (t,l[i+1])
m=deepcopy(x)#slope of (t[i+1],u[i+1])
b=deepcopy(x)# y cordinate of the function which is expected
x[0]=1 #intialization
b[0]=1
v[0]=0
h=t[1]-t[0] #step
for i in range(n-1):
  a[i]=-x[i] #slope condition given as m/k=1 
  k[i+1]=v[i]+(h*a[i]) #appliying heuns method twice
  l[i+1]=-(x[i]+(h*v[i]))
  v[i+1]=v[i]+(h/2)*(a[i]+l[i+1])
  u[i+1]=x[i]+(h*v[i])
  m[i+1]=v[i]+(h/2)*(a[i]+l[i+1])
  x[i+1]=x[i]+(h/2)*(v[i]+m[i+1])
for i in range(n-1):
  b[i+1]=b[0]*cos(t[i])#exact equation they have given
plot(t,b,color='red')#plot of exact values
print('solution:',x)#values we got
print("abs error :")
for i in range(n):print(abs(x[i]-b[i])) #absolute error
scatter(t,x,s=18)#plot of solution we got
# plot(t,v)
grid()
show()
```
